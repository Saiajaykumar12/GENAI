Prompt Templates - Dynamic Prompts

```python
from langchain_core.prompts import ChatPromptTemplate
llm = ChatOpenAI(model="gpt-3.5-turbo")
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role":"system","content":"You are a translator and translate input into {language}."},
    {"role":"user","content":"{query}"}
])
prompts
```

```python
final_prompts = prompts.format_messages(language="Hindi",query="Who is the president of India?")
```

```python
res2 = llm.invoke(final_prompts)
print(res2)
```

```python
res3 = llm.invoke(prompts.invkoke({"language":"Hindi","query":"Who is the president of India?"}))
print(res3.content)
```
