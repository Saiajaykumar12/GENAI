```python
from dotenv import load_dotenv
load_dotenv()
```

```python
documents = [
    "Machine Learnig is a subset of Artificial Intelligence",
    "Deep Leaning uses neural networks to model complex patterns in data",
    "Python is a popular programming language for data science and machine learning",
    "India won the Cricket World Cup in 2011",
    "AI models can understand and generate human-like text"
]
```

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
vector = embeddings.embed_documents(documents)
```

#### Store Vector Embeddings in Vector DB

```python
from langchain_chroma import Chroma
```

```python
vector_store = Chroma.from_texts(
    texts=documents,
    embedding=embeddings,
    persist_directory="/vector_db"
)
```

```python
query = "Machine Learning"
result = vector_store.similarity_search(query=query, k=2)
result
```
