```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_groq import ChatGroq
from langchain_chroma import Chroma
from langchain_core.prompts import ChatPromptTemplate
```

```python
loader = PyPDFLoader("data_science_syllabus.pdf")
docs = loader.load()
len(docs)
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splitted_data = splitter.split_documents(docs)
len(splitted_data)
```

```python
# Use this
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
```

```python
vector_store = Chroma.from_documents(
    documents=splitted_data,
    embedding=embeddings,
)
```

```python
query = "What is Machine Learning?"
data = vector_store.similarity_search(query=query)
```

```python
llm = ChatGroq(model="lama-3.3-70b-versatile")
```

```python
def get_context(query:str):
    data = vector_store.similarity_search(query=query)
    context = ""
    for doc in data:
        context += doc.page_content

    return {
        "context": context,
        "question": query
    }
```

```python
res = llm.invoke(f"""Can you provide me the answer based on provided context for my question
                 ,context:{context} and question:{query}""")
```

```python
prompt = PromptTemplate.from_template("""
You are a helpful assistant for answering question based on provided context.
                                      If you don't know the answer, say you don't know.
                                      Context: {context}
                                      Question: {question}""")
```

```python
rag_chain = get_context | prompt | llm
```
